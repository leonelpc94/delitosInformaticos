```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

dataset = pd.read_csv("Delitos_Informaticos_V1_20240924.csv")
dataset
```

```python
datos = dataset[["DEPARTAMENTO_HECHO","AÑO_HECHOS","TOTAL_PROCESOS"]]
datos20 = datos[datos["AÑO_HECHOS"]==2020]
datos21 = datos[datos["AÑO_HECHOS"]==2021]
datos22 = datos[datos["AÑO_HECHOS"]==2022]
datos23 = datos[datos["AÑO_HECHOS"]==2023]
a2020 = datos20["TOTAL_PROCESOS"].sum()
a2021 = datos21["TOTAL_PROCESOS"].sum()
a2022 = datos22["TOTAL_PROCESOS"].sum()
a2023 = datos23["TOTAL_PROCESOS"].sum()
listAnio =(2020,2021,2022,2023)
lista = (a2020,a2021,a2022,a2023)

lista

```

```python
cuartiles20 = datos20.groupby('DEPARTAMENTO_HECHO')['TOTAL_PROCESOS'].quantile([0.25,0.5,0.075]).unstack()
cuartiles20
```

```python
cuartiles21 = datos21.groupby('DEPARTAMENTO_HECHO')['TOTAL_PROCESOS'].quantile([0.25,0.5,0.075]).unstack()
cuartiles21
```

```python
cuartiles22 = datos22.groupby('DEPARTAMENTO_HECHO')['TOTAL_PROCESOS'].quantile([0.25,0.5,0.075]).unstack()
cuartiles22
```

```python
cuartiles23 = datos23.groupby('DEPARTAMENTO_HECHO')['TOTAL_PROCESOS'].quantile([0.25,0.5,0.075]).unstack()
cuartiles23
```

```python
#asignar valores a los ejes x e y 
x = listAnio
y = lista

#crear el grafico de barras 
bars = plt.bar(x,y, color= "#f51cde")
#agregar el valor a las barras
for bar in bars:
    yval = bar.get_height()  # Obtener la altura de la barra
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')  
#agregar un titulo al grafico
plt.title("Delitos infromaticos")
#agregar nombre a los ejes x e y
plt.xlabel("AÑO")
plt.ylabel("CANTIDAD DE DELITO")
#agregar valores especificos 
plt.xticks(x)
plt.yticks(y)
#agregar una grid en le eje y 
plt.grid(axis='y')
#ejecutar grafica 
plt.show()
```
